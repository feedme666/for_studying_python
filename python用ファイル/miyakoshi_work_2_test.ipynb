`a = np.pi`   
`b = np.e`

これにより、x には "$\pi$" が、yには "$e$" が代入できます。

## 課題2: 四則演算など
- 身長170cm、体重65kgの人のBMIを、変数"bmi"に代入してみましょう。
  - $BMI = 体重[kg]/(身長[m])^2$
- さらに、上記のBMIの自然対数($=log_e$)を、変数"log_bmi"に代入してみましょう。
  - Rは標準でも上記の計算ができますが、Pythonでは*math*や*numpy*などを読み込む必要があります。

***
## 回答チェック
以下、課題の回答が正解しているか確認するためのコードです。
実行してみましょう。


````python
import numpy as np
import math

def check_answer_1(max_numbers=None):
    original_max_numbers = max([1,2,3,5,8])
    try:
        if max_numbers is None:
            print('課題1で、"max_numbers" が定義されていません。')
        elif max_numbers == original_max_numbers:
            print('課題1は正解です。')
        else:
            print('課題1は不正解です。もう一度やり直しましょう')
    except NameError:
        print('課題1の、"max_numbers" が存在しません。変数名を確認してください。')

def check_answer_2(log_bmi=None):
    original_bmi = 65/((1.7)**2)
    original_log_from_numpy = np.log(original_bmi)
    original_log_from_math = math.log(original_bmi)
    try:
        if log_bmi is None:
            print('課題2で、"log_bmi" が定義されていません。')
        elif log_bmi == original_log_from_numpy or log_bmi == original_log_from_math:
            print('課題2は正解です。')
        else:
            print('課題2は不正解です。もう一度やり直しましょう')
    except NameError:
        print('課題2の、"log_bmi" が存在しません。変数名を確認してください。')

# テスト用の変数を定義していない状態でのチェック
try:
    check_answer_1(max_numbers)
except NameError:
    print('max_numbers が定義されていません。')

try:
    check_answer_2(log_bmi)
except NameError:
    print('log_bmi が定義されていません。')
````
